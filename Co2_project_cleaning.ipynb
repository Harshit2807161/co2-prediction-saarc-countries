```python
co2df['year'] = pd.to_datetime(co2df['year'], format='%Y')
```

```python
co2df = co2df.set_index('year')
```

```python
l = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Nepal',
       'Pakistan', 'Sri Lanka']
```

```python
afgh = co2df.loc[(co2df.country=='Afghanistan')].resample('M').ffill()
bang = co2df.loc[(co2df.country=='Bangladesh')].resample('M').ffill()
bhutan = co2df.loc[(co2df.country=='Bhutan')].resample('M').ffill()
india = co2df.loc[(co2df.country=='India')].resample('M').ffill()
nepal = co2df.loc[(co2df.country=='Nepal')].resample('M').ffill()
pakistan = co2df.loc[(co2df.country=='Pakistan')].resample('M').ffill()
srilanka = co2df.loc[(co2df.country=='Sri Lanka')].resample('M').ffill()
```

```python
final_monthly_data = pd.concat([afgh,bang,bhutan,india,nepal,pakistan,srilanka])
```

```python
final_monthly_data.to_csv("input_data_clean_monthly.csv")
```
